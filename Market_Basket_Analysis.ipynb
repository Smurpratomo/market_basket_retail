### 1. Total Customers by Nations

```python
temp1= df.groupby(['Customer ID','Country'])['Customer ID'].size().unstack()
temp1.loc['Total',:]=temp1.count(axis=0)
```

```python
temp1
```

### 2. Total Orders by Nations

```python
temp2= df.groupby(['Invoice','Country'])['Invoice'].size().unstack()
temp2.loc['Total',:]=temp2.count()
```

```python
temp2
```

### 3. Combining

```python
temp1_=temp1.loc['Total',:]
df1=pd.DataFrame(data=temp1_)
df1=df1.sort_values(by=['Total'], ascending=False)

df1.reset_index(level=0, inplace=True)
df1.rename(columns={'Total':'Total Customers'}, inplace=True)
df1.head()
```

```python
temp2_=temp2.loc['Total',:]
df2=pd.DataFrame(data=temp2_)
df2=df2.sort_values(by=['Total'], ascending=False)

df2.reset_index(level=0, inplace=True)
df2.rename(columns={'Total':'Total Orders'}, inplace=True)
df2.head()
```

```python
df3 = df1.merge(df2)
df3['Orders per Customer'] = (df3['Total Orders']/df3['Total Customers']).round(decimals=2)
df3=df3.sort_values(by=['Orders per Customer'], ascending=False)
df3.reset_index(drop=True, inplace=True)
df3= df3.loc[:9,:]
df3
```

### 4. Sweden

```python
df = df.loc[df.Country == 'Sweden']
```

```python
df.head()
```

```python
temp3= df.groupby('Invoice')['Description'].apply(lambda x: ','.join((x))).reset_index()
temp3
```

```python
transactions = temp3.Description.str.split(",")
```

```python
transactions
```

```python
from mlxtend.preprocessing import TransactionEncoder
encoder= TransactionEncoder()
onehot= encoder.fit(transactions).transform(transactions)
onehot= pd.DataFrame(data= onehot, columns= encoder.columns_)
```

```python
onehot.mean()
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
frequent_items= apriori(onehot, min_support= 0.05, use_colnames=True, max_len= 2)
rules= association_rules(frequent_items, metric='support', min_threshold= 0.1)
```

```python
rules['antecedents'] = rules['antecedents'].apply(lambda x: ','.join(list(x)))
rules['consequents'] = rules['consequents'].apply(lambda x: ','.join(list(x)))
```

```python
rules
```

```python
import seaborn as sns
table = rules.pivot(index='antecedents', columns='consequents', values='support')
sns.heatmap(table, annot=True, cmap='coolwarm', vmin= 0, vmax=1)
```
